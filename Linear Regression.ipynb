```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y ,test_size=0.2) # 20 -30
```

```python
print(x_train.shape)
print(x_test.shape)
print()
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
```

```python
lm.fit(x_train,y_train)
```
